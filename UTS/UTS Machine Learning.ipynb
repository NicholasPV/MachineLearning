Machine Learning adalah cabang dari kecerdasan buatan (Artificial Intelligence) yang fokus pada pengembangan algoritma dan teknik yang memungkinkan komputer untuk "belajar" dari dan membuat prediksi atau keputusan berdasarkan data. Secara sederhana, machine learning memungkinkan sistem komputer untuk meningkatkan kinerja mereka pada tugas-tugas tertentu melalui pengalaman tanpa perlu diprogram secara eksplisit untuk setiap tugas tersebut.

2. Penerapan, Alasan Dibutuhkan, dan Manfaat Machine Learning

Contoh penerapan Machine Learning dalam kehidupan sehari-hari adalah sebagai berikut:
- Pengenalan Wajah (Face Recognition), yaitu penggunaan pengenalan wajah di smartphone untuk membuka kunci perangkat atau mengautentikasi pengguna untuk berbagai aplikasi.
- Sistem Rekomendasi (Recommendation Systems), yaitu rekomendasi produk di platform e-commerce seperti Tokopedia, rekomendasi video di YouTube, rekomendasi film di Netflix, rekomendasi musik di Spotify, dan lain-lain.
- Navigasi dan Transportasi (Navigation and Transportation), yaitu aplikasi navigasi seperti Google Maps yang menggunakan machine learning untuk memberikan rute terbaik dan estimasi waktu tiba.
- Asisten Virtual (Virtual Assistants), yaitu asisten virtual seperti Siri, Google Assistant, atau Alexa yang dapat memahami dan merespons perintah suara.
- Kendaraan Otonom (Autonomous Vehicles), yaitu Mobil self-driving yang menggunakan machine learning untuk memahami lingkungan mereka dan membuat keputusan mengemudi, seperti Tesla.

Machine learning dibutuhkan karena dapat mengotomatisasi dan meningkatkan proses pengambilan keputusan, mengenali pola dalam data yang sangat besar, dan memberikan solusi yang lebih cepat dan efisien dibandingkan dengan metode tradisional. Hal ini memungkinkan berbagai industri untuk berinovasi, meningkatkan efisiensi, dan memberikan layanan yang lebih baik kepada pengguna

Machine Learning memberikan berbagai manfaat, seperti berikut:
- Peningkatan Efisiensi: Mengotomatisasi tugas-tugas yang berulang dan kompleks.
- Kecepatan dan Akurasi: Membuat prediksi dan keputusan dengan cepat dan akurat.
- Personalized Experience: Memberikan pengalaman yang lebih dipersonalisasi kepada pengguna.
- Keamanan: Meningkatkan keamanan di berbagai aplikasi seperti deteksi penipuan dan keamanan siber.
- Inovasi: Memungkinkan pengembangan teknologi baru dan solusi inovatif.

3. Taxonomi dalam penerapan Machine Learning mengacu pada berbagai kategori dan jenis pembelajaran mesin berdasarkan cara model dilatih dan tipe masalah yang dipecahkan. Berikut adalah beberapa macam taxonomi dalam machine learning:
- Supervised Learning, yaitu taxonomi yang melatih model menggunakan data yang diberi label. Data input memiliki output yang sesuai, dan model belajar untuk memetakan input ke output yang benar.
- Unsupervised Learning, yaitu taxonomi yang melatih model menggunakan data yang tidak diberi label. Algoritma mencoba menemukan struktur atau pola tersembunyi dalam data.
- Semi-Supervised Learning, yaitu taxonomi yang menggabungkan data berlabel dan tidak berlabel dalam proses pelatihan. Ini berguna ketika pelabelan data sangat mahal atau memakan waktu.
- Self-Supervised Learning, yaitu taxonomi yang melatih model untuk memprediksi sebagian dari data berdasarkan sebagian lainnya. Ini sering digunakan untuk pretraining model dengan data yang tidak berlabel sebelum fine-tuning dengan data berlabel.
- Reinforcement Learning, yaitu taxonomi yang dimana agen belajar melalui interaksi dengan lingkungan. Agen menerima reward (penghargaan) atau punishment (hukuman) berdasarkan tindakan yang diambil, dan tujuan agen adalah memaksimalkan total reward.

## No. 1
Sempurnakan Data Set Universitas Esa Unggul memiliki detail mahasiswa sebagai berikut :
Ani, Budi, Joni, Jono dan Lono Setiap mahasiswa dating ke kampus mulai dari hari senin sampai minggu Untuk melakukan kegiatan kuliah Ani menghabiskan biaya 30000 dikampus setiap datang Budi menghabiskan biaya 35000 dikampus setiap datang Joni menghabiskan biaya 20000 dikampus setiap datang Jono 25000 dan Lono 15000.

a) Berapa rata-rata mahasiswa datang pada minggu ini? <br>
b) Kapan biaya tertinggi terjadi? <br>
c) Hari apa biaya lebih dari 110000? <br>
d) Siapa yang paling banyak datang ke kampus? <br>
e) Siapa yang datang pada hari minggu? <br>
f) Berapa biaya tertinggi dan terendah? <br>
g) Berapa frekuensi datang tertinggi dan terendah?

````python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Import Dataset
dataset = pd.read_csv('DatasetNo1.csv')
print(dataset)
````

````python
# a) Berapa rata-rata mahasiswa datang pada minggu ini?
meanDatang = dataset['Datang'].mean()
print("Rata-rata mahasiswa yang datang pada minggu ini adalah sebanyak: ", meanDatang)
````

````python
# b) Kapan biaya tertinggi terjadi?
dataset['Total Biaya'] = dataset['Datang'] * dataset['Biaya']
maxBiaya = dataset.loc[dataset['Total Biaya'].idxmax()]
print("Biaya tertinggi terjadi pada:\n", maxBiaya)
````

````python
# c) Hari apa biaya lebih dari 110000?
dataset['Total Biaya'] = dataset['Datang'] * dataset['Biaya']
biggerThan110000Biaya = dataset[dataset['Total Biaya'] > 110000]['Hari'].values
print("Hari dengan biaya yang melebihi 110000 adalah: ", biggerThan110000Biaya)
````

````python
# d) Siapa yang paling banyak datang ke kampus?
modusMahasiswa = dataset['Mahasiswa'].value_counts().idxmax()
print("Mahasiswa yang paling banyak datang ke kampus adalah: ", modusMahasiswa)
````

````python
# e) Siapa yang datang pada hari minggu?
sundayMahasiswa = dataset[dataset['Hari'] == 'Minggu']['Mahasiswa'].values
print("Mahasiswa yang datang pada hari Minggu adalah: ", sundayMahasiswa)
````

````python
# f) Berapa biaya tertinggi dan terendah?
dataset['Total Biaya'] = dataset['Datang'] * dataset['Biaya']
highestBiaya = dataset['Total Biaya'].max()
lowestBiaya = dataset['Total Biaya'].min()
print("Biaya tertinggi adalah sebanyak: ", highestBiaya)
print("Biaya terendah adalah sebanyak: ", lowestBiaya)
````

````python
# g) Berapa frekuensi datang tertinggi dan terendah?
highestDatang = dataset['Datang'].max()
lowestDatang = dataset['Datang'].min()
print("Frekuensi datang tertinggi adalah sebanyak: ", highestDatang)
print("Frekuensi datang terendah adalah sebanyak: ", lowestDatang)
````

Sehingga dapat menghasilkan seperti berikut

````python
import pandas as pd
from plotnine import ggplot, aes, geom_bar, labs, theme_minimal, theme, element_text, scale_fill_manual

# Membuat data frame info_mahasiswa
data = {
    'fakultas': ["Bisnis", "D3 Perhotelan", "ICT", "Ilmu Komunikasi", "Seni dan Desain"],
    'jumlah_mahasiswa': [260, 28, 284, 465, 735],
    'akreditasi': ["A", "A", "B", "A", "A"]
}
info_mahasiswa = pd.DataFrame(data)

# Membuat grafik batang menggunakan plotnine
colors = ['#FF6F61', '#6B8E23', '#20B2AA', '#6495ED', '#DA70D6']

gambar = (
    ggplot(info_mahasiswa, aes(x='fakultas', y='jumlah_mahasiswa', fill='fakultas')) +
    geom_bar(stat="identity", width=0.7) +
    scale_fill_manual(values=colors) +
    labs(title="Jumlah Mahasiswa per Fakultas", x="Fakultas", y="Jumlah Mahasiswa") +
    theme_minimal() +
    theme(axis_text_x=element_text(angle=45, hjust=1))
)

# Menampilkan grafik
print(gambar)
````
